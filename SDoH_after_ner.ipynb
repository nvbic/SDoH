```python
from pathlib import Path
import pickle as pkl
import numpy as np
import os
```

```python
def pkl_save(data, file):
    with open(file, "wb") as f:
        pkl.dump(data, f)
        
def pkl_load(file):
    with open(file, "rb") as f:
        data = pkl.load(f)
    return data
```

```python
data_root='/home/zehao.yu/workspace/py3/SDoH/res/'
```

```python
dict1=pkl_load(data_root+'breast.pkl')
dict2=pkl_load(data_root+'colo.pkl')
dict3=pkl_load(data_root+'lung.pkl')
```

```python

def find_agg_data(dict1):
    null_person=set()
    patient=set()
    dict_agg=dict()
    for k,v in dict1.items():
        if len(v)==0:
            null_person.add(k)
        else:
            patient.add(k)
            for k1,v1 in v.items():
                if k1 not in dict_agg.keys():
                    dict_agg.update({k1:len(v1)})
                else:
                    dict_agg[k1]+=len(v1)
    return patient,dict_agg
    #break
```

```python
#find pts in patient level
def find_agg_data_2(dict1):
    null_person=set()
    patient=set()
    dict_agg=dict()
    for k,v in dict1.items():
        if len(v)==0:
            null_person.add(k)
        else:
            patient.add(k)
            for k1,v1 in v.items():
                if k1 not in dict_agg.keys():
                    dict_agg.update({k1:set()})
                    dict_agg[k1].add(k)
                else:
                    dict_agg[k1].add(k)
    return patient,dict_agg
```

```python
bc_pts,bc_dict_agg=find_agg_data(dict1)
cc_pts,cc_dict_agg=find_agg_data(dict2)
lc_pts,lc_dict_agg=find_agg_data(dict3)
```

```python
bc_pts,bc_dict_agg=find_agg_data_2(dict1)
cc_pts,cc_dict_agg=find_agg_data_2(dict2)
lc_pts,lc_dict_agg=find_agg_data_2(dict3)
```

```python
data={'SDoH_cate':sorted(list(bc_dict_agg.keys()))}
df=pd.DataFrame(data)
```

```python
def count_pts(x,dict_agg):
    return len(dict_agg[x])
def ave_pts(x,dict_agg,pts):
    return len(dict_agg[x])/len(pts)
```

```python
df['bc_count']=df.apply(lambda x: count_pts(x['SDoH_cate'],bc_dict_agg),axis=1)
df['bc_ave']=df.apply(lambda x: ave_pts(x['SDoH_cate'],bc_dict_agg,bc_pts),axis=1)
df['cc_count']=df.apply(lambda x: count_pts(x['SDoH_cate'],cc_dict_agg),axis=1)
df['cc_ave']=df.apply(lambda x: ave_pts(x['SDoH_cate'],cc_dict_agg,cc_pts),axis=1)
df['lc_count']=df.apply(lambda x: count_pts(x['SDoH_cate'],lc_dict_agg),axis=1)
df['lc_ave']=df.apply(lambda x: ave_pts(x['SDoH_cate'],lc_dict_agg,lc_pts),axis=1)
```
